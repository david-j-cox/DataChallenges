# The Clairvoyant Constructor

# Goals
You are a data scientist at a large construction company. To submit more competitive bids for contracts, the head of analytics and data science would like you to forecast core construction material costs. You have access to the pricing information (per unit, in dollars) for the company’s past purchases of plywood, sheetrock, steel beams, rebar, and glass contracts over the past ten years in csv format. 

Currently, the company can purchase futures contracts in a material for up to six months. Thus, your manager is most interested in forecasting the price of each commodity up to six months into the future. In particular, your manager feels it would be really useful to understand when a commodity price is about to increase temporarily, so the company can buy the commodity early or wait until after the temporary increase. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import sys
from pandas_profiling import ProfileReport
```

```python
# Read in the data for EDA. 
data = pd.read_csv('construction_material_prices_2008_2018.csv', low_memory=False)
df = data.copy()
df_cols = list(df)
```

```python
profile = ProfileReport(df)
profile
```

```python
# Change the date col from category to datetime type
df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'], infer_datetime_format=True)
```

```python
df.dtypes
```

```python
# Take a look at what the time-series look like visually
df_plot = df.set_index('Unnamed: 0')
df_plot.plot(subplots=True, figsize=(15, 10))
plt.xlabel('Date', fontsize=20)
```

```python
# Same visualization as above using interpolated vals
df_plot_cols = list(df_plot)
df_interpol = df_plot.resample('W').mean()
for i in df_plot_cols:
    df_interpol[i] = df_interpol[i].interpolate(method='time')
df_interpol = df_interpol.dropna()
df_interpol.plot(subplots=True, figsize=(15, 10))
plt.xlabel('Date', fontsize=20)
```

# Forecasting Models
1. Autoregression - good with data without trends or seasonal components
1. Moving average - good with data without trends or seasonal components
1. Autoregressive Moving Average (ARMA) - good with data without trends or seasonal components
1. Autoregressive Integrated Moving Average (ARIMA) - good with data with trend and without seasonal components
1. Seasonal Autoregressive Integreated Moving-Average (SARIMA) - good with data with trend and/or seasonal components
1. Seasonal Autoregressive Integrated Moving-Average with Exogenous Regressors (SARIMAX)
1. Vector Autoregression (VAR)
1. Vector Autoregression Moving-Average (VARMA)
1. Vector Autoregression Moving-Average with Exogenous Regressors (VARMAX)
1. Simple Exponential Smoothing (SES)
1. Holt Winter's Exponential Smoothing (HWES)
1. Prophet

```python
# Check out autocorrelation over 44 weeks (6.5 months)
from statsmodels.tsa.stattools import acf
for i in df_plot_cols:
    lac_acf = acf(df_interpol[i], nlags=44, fft=False)
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    plt.stem(lac_acf)
    plt.axhline(y=0, linestyle='-', color='black')
    plt.xlabel('Lag', fontsize=20)
    plt.ylabel('ACF', fontsize=20)
    plt.title(i, fontsize=20)
    plt.show()
```

### It looks like there's some correlation between prices of each commodity around 180 days out. But, the correlations are relatively weak so probably not helpful to use this alone. 

#### Decompose time series 

```python
import statsmodels.api as sm
from pylab import rcParams

for i in df_plot_cols:
    data = df_interpol[i]
    data_decomp = sm.tsa.seasonal_decompose(data, model='additive')
    fig = data_decomp.plot()
    plt.xlabel(i, fontsize=20)
    plt.show
```

### All five commodities show fairly clear seasonality. Let's replot the same data but zooming in on a two-year span to see how the six-month time frames change. 

```python
import statsmodels.api as sm
from pylab import rcParams

for i in df_plot_cols:
    data = df_interpol[i][-120:]
    data_decomp = sm.tsa.seasonal_decompose(data, model='additive')
    fig = data_decomp.plot()
    plt.xlabel(i, fontsize=20)
    plt.show
```

## SARIMA

```python
# Data with missing vals
from statsmodels.tsa.statespace.sarimax import SARIMAX
for i in df_plot_cols:
    model = SARIMAX(df_interpol[i], order=(1, 1, 1), seasonal_order=(1, 1, 1, 1))
    model_fit = model.fit(disp=False)
    y_pred = model_fit.predict(len(df_interpol), len(df_interpol))
    print(y_pred)
```
